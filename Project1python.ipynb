**Exercise 1**: Write a function count_vowels(s) that counts the number of vowels (a, e, i, o, u) in a given string 'This Is Just A String That We Use To Practice'


**Yêu cầu:** viểt function count.vowels(s)

**Input**: string

**Output:** number


# **Hướng xử lý**:


Duyệt từng ký tự trong string -> loop kiểm tra nguyên âm -> condition: đúng thì +1 biến đếm


#**Logic hoàn chỉnh:**

function count_vowels(s):

convert s thành lowercase

tạo biến count = 0 (1 hộp rỗng tên count)

for mỗi kí tự trong s:

  nếu ký tự thuộc (ueoai):

  -> count = count + 1


```python
#Cách 1:
def count_vowels(s):
  s = s.lower()
  count = 0

  for char in s:
    if char in "aeiou":
      count += 1
  return count

text = "This Is Just A String That We Use To Practice"
print(count_vowels(text))


#cách 2:
def count_vowels(s):
  count = 0

  for char in s.lower():
    if char in "ueoai":
      count +=1

  return count

count_vowels("This Is Just A String That We Use To Practice")
```

**Đề 2a:** You are managing a small library system and have a list of books. Each book has a title, author, and whether it is currently borrowed. Your task is to organize and analyze the data.

**Input:** A list of dictionaries, where each dictionary contains information about a book:

books = [ {"title": "1984", "author": "George Orwell", "borrowed": True},

 {"title": "To Kill a Mockingbird", "author": "Harper Lee", "borrowed": False},

{"title": "The Great Gatsby", "author": "F. Scott Fitzgerald", "borrowed": True},

{"title": "Moby Dick", "author": "Herman Melville", "borrowed": False},

{"title": "War and Peace", "author": "Leo Tolstoy", "borrowed": True}, ]

Write a function that takes ***a title as input*** and checks if the book exists in the library. If it does, return 'Borrowed' else return 'Available'. If it doesn't return 'Not exist in the library'

**Output mẫu:**

'Borrowed' # với item = '1984'

'Available' # với item = 'To Kill a Mockingbird'

'Not exist in the library' # với item = 'A'

# hướng xử lý:
đây là list dictionary, ta truy cập = key (book["tittle"]).

Input = tittle (tên sách)

Nếu sách tồn tại:

- Borrowed = True -> borrowed
- Borrowed = False -> available

Nếu ko tìm thấy -> Not exist in the library

#LOGIC BREAKDOWN
- Check qua từng book trong list
- So sánh tittle của book với input
- Nếu trùng -> kiểm tra Borrowed
- Nếu duyệt ko có -> not exist


```python
#Cách 1:

def check_book(title):
  for book in books:
    if book["title"] == title:
      if book["borrowed"]:
        return "Borrowed"
      else:
        return "Available"
  return "Not exist in the library"

books = [ {"title": "1984", "author": "George Orwell", "borrowed": True},
{"title": "To Kill a Mockingbird", "author": "Harper Lee", "borrowed": False},
{"title": "The Great Gatsby", "author": "F. Scott Fitzgerald", "borrowed": True},
{"title": "Moby Dick", "author": "Herman Melville", "borrowed": False},
{"title": "War and Peace", "author": "Leo Tolstoy", "borrowed": True}, ]

print(check_book("1984"))
print(check_book("To Kill a Mockingbird"))
print(check_book ("A"))


#Cách 2 Pythonic

def check_book(title):
  for book in books:
    if book["title"] == title:
      return "Borrowed" if book["borrowed"] else "Available"
  return "Not exist in the library"


print(check_book("1984"))
print(check_book("To Kill a Mockingbird"))
print(check_book ("A"))
```

**Exercise 2b:** Count Borrowed and Available Books from books in previous exercise

**Input:**

**Output:** {"borrowed": 3, "available": 2}

Hướng xử lý & breakdown


def function

Tạo hộp count rỗng

For mỗi tittle trong book, nếu borrowed & available

-> count = count + 1

```python
#Cách 1 (Fixed):

books = [ {"title": "1984", "author": "George Orwell", "borrowed": True},
{"title": "To Kill a Mockingbird", "author": "Harper Lee", "borrowed": False},
{"title": "The Great Gatsby", "author": "F. Scott Fitzgerald", "borrowed": True},
{"title": "Moby Dick", "author": "Herman Melville", "borrowed": False},
{"title": "War and Peace", "author": "Leo Tolstoy", "borrowed": True}, ]

def count_books():
  borrowed_count = 0
  available_count = 0
  for book in books:
    if book["borrowed"]:
      borrowed_count += 1
    else:
      available_count += 1
  return {"borrowed": borrowed_count, "available": available_count}

print(count_books())
```
